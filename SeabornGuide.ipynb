```python
import seaborn as sns
%matplotlib inline
```

```python
# We will initially see Distribution plots
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distribution Plots
"""This will give us the distribution and the histogram
set kde = false to avoid the distribution line """

sns.distplot(tips['total_bill'],kde= False,bins = 35)
```

```python
# Joint Plot 

sns.jointplot(x='total_bill' ,y='tip',data= tips,kind = "kde")

"""change the attribute kind to get different kinds of attributes"""
```

```python
#pairplot

sns.pairplot(tips,hue="sex",palette ="coolwarm")
```

```python
# rugplots in seaborn

sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
tips.head()
```

```python
#BarPlot
#aggregates fucntion as mean  by default   
import numpy as np
sns.barplot(x = 'sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x = "sex",data = tips)
```

```python
#BoxPlots

sns.boxplot(x = 'day',y='total_bill',data=tips,hue="smoker")
```

```python
#Violin Plot
sns.violinplot(x = 'day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# strip plot - it draws a scatter plot x = categorical and y = numerical

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge = True)
```

```python
#swarmplot - it is like a stripplot but doesnt allow points to overlap
# combination of the Violin plot and Strip plot
#drawback = many times they  dont scale well and cannot give good viewing result
#sns.violinplot(x = 'day',y='total_bill',data=tips)
sns.swarmplot(x = 'day',y='total_bill',data=tips)
```

```python
#Factor Plot 

sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
