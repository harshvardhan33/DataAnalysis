```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Ye hai Title")
plt.show()
```

```python
# Creating Multiple Plots
plt.subplot(1,2,1)
plt.xlabel("X value Red")
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.xlabel("X value Blues")
plt.plot(y,x,'b')
```

```python
# Object Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("ye toh x hai ")
axes.set_ylabel("ye toh y hai ")
axes.set_title("ye toh title hai ")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.5,0.3,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols =2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="x and y")
ax.plot(y,x,label="y and x")
ax.legend(loc=0)
```
